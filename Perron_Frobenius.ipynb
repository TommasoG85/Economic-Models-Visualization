```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define your production matrix A
A = np.array([[0.5, 0.1, 0.5],
              [0.3, 0.4, 0.3],
              [0.2, 0.1, 0.4]])

# Calculate eigenvalues and eigenvectors
eigenvalues, eigenvectors = np.linalg.eig(A)

# Find the dominant eigenvalue and its corresponding eigenvector
dominant_index = np.argmax(eigenvalues)
dominant_eigenvalue = eigenvalues[dominant_index]
dominant_eigenvector = eigenvectors[:, dominant_index]

# Normalize the dominant eigenvector to represent proportions
growth_vector = dominant_eigenvector / np.sum(dominant_eigenvector)

# Create a new figure for the 3D plot
fig = plt.figure(figsize=(10, 12))
ax = fig.add_subplot(111, projection='3d')

# Set the axes limits
max_length = 0.5
ax.set_xlim([0, max_length])
ax.set_ylim([0, max_length])
ax.set_zlim([0, max_length])

# Add projections for the original matrix values
for i in range(3):  # Loop through each column (sector) of matrix A
    ax.plot([A[0, i], A[0, i]], [0, 0], [0, A[2, i]], color='gray', linestyle='--', linewidth=2)  # X-axis projection
    ax.plot([0, 0], [A[1, i], A[1, i]], [0, A[2, i]], color='gray', linestyle='--', linewidth=2)  # Y-axis projection
    ax.plot([0, 0], [0, 0], [A[2, i], A[2, i]], color='gray', linestyle='--', linewidth=2)  # Z-axis projection

    # Add connecting dashed lines
    ax.plot([A[0, i], A[0, i]], [A[1, i], 0], [A[2, i], A[2, i]], color='gray', linestyle='--', linewidth=2)
    ax.plot([A[0, i], 0], [A[1, i], A[1, i]], [A[2, i], A[2, i]], color='gray', linestyle='--', linewidth=2)
    ax.plot([A[0, i], A[0, i]], [A[1, i], A[1, i]], [A[2, i], 0], color='gray', linestyle='--', linewidth=2)

# Add the three vectors
for i in range(3):  # Loop through each column (sector) of matrix A
    ax.quiver(0, 0, 0, A[0, i], A[1, i], A[2, i],
              color='black', linestyle='--', linewidth=2, arrow_length_ratio=0.1)

# Draw the growth vector (red)
ax.quiver(0, 0, 0, growth_vector[0], growth_vector[1], growth_vector[2],
          color='red', arrow_length_ratio=0.1, linewidth=3, label='Autovettore dominante (Crescita massima)')

# Draw projections onto each axis
# X-axis projection
ax.plot([growth_vector[0], growth_vector[0]], [0, 0], [0, growth_vector[2]], color='orange', linestyle='--', linewidth=3)
# Y-axis projection
ax.plot([0, 0], [growth_vector[1], growth_vector[1]], [0, growth_vector[2]], color='green', linestyle='--', linewidth=3)
# Z-axis projection
ax.plot([0, 0], [0, 0], [growth_vector[2], growth_vector[2]], color='blue', linestyle='--', linewidth=3)


# Draw connecting lines from vector endpoint to projections
ax.plot([growth_vector[0], growth_vector[0]], [growth_vector[1], 0], [growth_vector[2], growth_vector[2]], color='orange', linestyle='--', linewidth=3)  # Connection to x-axis projection
ax.plot([growth_vector[0], 0], [growth_vector[1], growth_vector[1]], [growth_vector[2], growth_vector[2]], color='green', linestyle='--', linewidth=3)  # Connection to y-axis projection
ax.plot([growth_vector[0], growth_vector[0]], [growth_vector[1], growth_vector[1]], [growth_vector[2], 0], color='blue', linestyle='--', linewidth=3)  # Connection to z-axis projection

# Adding labels and title
ax.set_xlabel('Settore 1')
ax.set_ylabel('Settore 2')
ax.set_zlabel('Settore 3')
ax.set_title('Vettore di crescita nel modello di Von Neumann', fontsize=16)
ax.legend()

# Show the plot
plt.grid(False)
plt.tight_layout()
plt.show()



import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the technical coefficients matrix for the Sraffa model
A = np.array([[0.5, 0.2, 0.3],
              [0.1, 0.4, 0.5],
              [0.3, 0.3, 0.4]])

# Calculate eigenvalues and eigenvectors
eigenvalues, eigenvectors = np.linalg.eig(A)

# Find the dominant eigenvalue and its corresponding eigenvector
dominant_index = np.argmax(eigenvalues)
dominant_eigenvalue = eigenvalues[dominant_index]
dominant_eigenvector = eigenvectors[:, dominant_index]

# Normalize the dominant eigenvector to represent proportions
normalized_vector = dominant_eigenvector / np.sum(dominant_eigenvector)

# Create a new figure for the 3D plot
fig = plt.figure(figsize=(10, 12))
ax = fig.add_subplot(111, projection='3d')

# Set the axes limits
max_length = 0.5
ax.set_xlim([0, max_length])
ax.set_ylim([0, max_length])
ax.set_zlim([0, max_length])

# Draw the normalized vector (red) representing technical coefficients
ax.quiver(0, 0, 0, normalized_vector[0], normalized_vector[1], normalized_vector[2],
          color='red', arrow_length_ratio=0.1, linewidth=3, label='Autovettore dominante (Proporzioni Tecniche)')

# Add projections for the original matrix values
for i in range(3):  # Loop through each column (sector) of matrix A
    ax.plot([A[0, i], A[0, i]], [0, 0], [0, A[2, i]], color='gray', linestyle='--', linewidth=2)  # X-axis projection
    ax.plot([0, 0], [A[1, i], A[1, i]], [0, A[2, i]], color='gray', linestyle='--', linewidth=2)  # Y-axis projection
    ax.plot([0, 0], [0, 0], [A[2, i], A[2, i]], color='gray', linestyle='--', linewidth=2)  # Z-axis projection

    # Add connecting dashed lines
    ax.plot([A[0, i], A[0, i]], [A[1, i], 0], [A[2, i], A[2, i]], color='gray', linestyle='--', linewidth=2)
    ax.plot([A[0, i], 0], [A[1, i], A[1, i]], [A[2, i], A[2, i]], color='gray', linestyle='--', linewidth=2)
    ax.plot([A[0, i], A[0, i]], [A[1, i], A[1, i]], [A[2, i], 0], color='gray', linestyle='--', linewidth=2)

# Add the three vectors
for i in range(3):  # Loop through each column (sector) of matrix A
    ax.quiver(0, 0, 0, A[0, i], A[1, i], A[2, i],
              color='black', linestyle='--', linewidth=2, arrow_length_ratio=0.1)

# Draw projections onto each axis
# X-axis projection
ax.plot([normalized_vector[0], normalized_vector[0]], [0, 0], [0, normalized_vector[2]], color='orange', linestyle='--', linewidth=3)
# Y-axis projection
ax.plot([0, 0], [normalized_vector[1], normalized_vector[1]], [0, normalized_vector[2]], color='green', linestyle='--', linewidth=3)
# Z-axis projection
ax.plot([0, 0], [0, 0], [normalized_vector[2], normalized_vector[2]], color='blue', linestyle='--', linewidth=3)

# Draw connecting lines from vector endpoint to projections
ax.plot([normalized_vector[0], normalized_vector[0]], [normalized_vector[1], 0], [normalized_vector[2], normalized_vector[2]], color='orange', linestyle='--', linewidth=3)  # Connection to x-axis projection
ax.plot([normalized_vector[0], 0], [normalized_vector[1], normalized_vector[1]], [normalized_vector[2], normalized_vector[2]], color='green', linestyle='--', linewidth=3)  # Connection to y-axis projection
ax.plot([normalized_vector[0], normalized_vector[0]], [normalized_vector[1], normalized_vector[1]], [normalized_vector[2], 0], color='blue', linestyle='--', linewidth=3)  # Connection to z-axis projection

# Adding labels and title
ax.set_xlabel('Settore 1')
ax.set_ylabel('Settore 2')
ax.set_zlabel('Settore 3')
ax.set_title('Coefficienti Tecnici nel Modello di Sraffa', fontsize=16)
ax.legend()

# Show the plot
plt.grid(False)
plt.tight_layout()
plt.show()

import numpy as np
import networkx as nx
import matplotlib.pyplot as plt

# Create a directed graph
G = nx.DiGraph()

# Add edges to the graph (example three-node graph)
G.add_edge(0, 1)
G.add_edge(1, 2)
G.add_edge(2, 0)
G.add_edge(1, 0)

# Calculate PageRank values
pagerank = nx.pagerank(G, alpha=0.85)

# Get PageRank values for node sizes
node_sizes = [pagerank[node] * 3000 for node in G]

# Get node colors based on PageRank values
node_colors = [pagerank[node] for node in G]

# Draw the graph
pos = nx.spring_layout(G)  # Position the nodes using a spring layout
fig, ax = plt.subplots(figsize=(10, 8))

nx.draw_networkx_edges(G, pos, arrowstyle='->', arrowsize=20, edge_color='gray', ax=ax)
nx.draw_networkx_nodes(G, pos, node_size=node_sizes, node_color=node_colors, cmap=plt.cm.viridis, alpha=0.9, ax=ax)
nx.draw_networkx_labels(G, pos, font_size=12, font_color='black', ax=ax)

# Add a color bar to represent the PageRank scale
sm = plt.cm.ScalarMappable(cmap=plt.cm.viridis, norm=plt.Normalize(vmin=min(pagerank.values()), vmax=max(pagerank.values())))
sm.set_array([])
cbar = plt.colorbar(sm, ax=ax, label='PageRank Score')

# Title and display
ax.set_title('2D PageRank Visualization of a Three-Node Graph')
ax.axis('off')
plt.show()







```
